```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python

# Assuming the file is named 'medical_sales_data.csv'
df = pd.read_csv('medical-sales-data.csv')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Aggregate the data by date and medicine
df['ds'] = pd.to_datetime(df[['year', 'month', 'day']])
df_grouped = df.groupby(['ds', 'medicine_name'])['total_sales'].sum().reset_index()

# Display the grouped data
df_grouped.head()
```

```python
# Initialize a dictionary to store models and forecasts for each medicine
models = {}
forecasts = {}

# Get the list of unique medicines
medicines = df_grouped['medicine_name'].unique()

# Fit the model for each medicine
for medicine in medicines:
    # Filter the data for the current medicine
    df_medicine = df_grouped[df_grouped['medicine_name'] == medicine]
    df_medicine = df_medicine[['ds', 'total_sales']].rename(columns={'total_sales': 'y'})

    # Initialize and fit the Prophet model
    model = Prophet()
    model.fit(df_medicine)

    # Create a DataFrame with future dates
    future = model.make_future_dataframe(periods=30)  # Forecast for the next 30 days

    # Make predictions
    forecast = model.predict(future)

    # Store the model and forecast
    models[medicine] = model
    forecasts[medicine] = forecast

    # Plot the forecast
    fig1 = model.plot(forecast)
    plt.title(f'Sales Forecast for {medicine}')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.show()

    # Plot the forecast components
    fig2 = model.plot_components(forecast)
    plt.show()
```
